```python
# construct graph from the given data
graph = {}

for i in range(1,len(arr)):
  for j in range(1,len(arr[i])):
    c = arr[i][j]
    if(c=='-'):
      c = 0
    start=arr[i][0].lower()
    end = arr[0][j].lower()

    if start not in graph:
      graph[start] = {}
    graph[start][end] = int(c)
    if end not in graph:
      graph[end] = {}
    graph[end][start] = int(c)

print(len(graph))
```

```python
import sys

def find_minnode(dist, visited):
  mindist = sys.maxsize
  min_node = 'none'
  for node in dist.keys():
    if dist[node] < mindist and visited[node] == False:
      mindist = dist[node]
      min_node = node
  return min_node

def dijkstra(start,graph,nodes):
  total_node = len(nodes)
  dist = {}
  visited = {}
  for i in nodes:
    visited[i] = False
    dist[i] = sys.maxsize
  dist[start] = 0

  for i in range(total_node):
    u = find_minnode(dist,visited)
    if u=='none':
      return dist
    visited[u] = True

    for v in graph[u].keys():
      if graph[u][v]>0 and visited[v]==False and dist[v]>dist[u]+graph[u][v]:
        dist[v] = dist[u]+graph[u][v]

  return dist
```

```python
nodes = []
for i in range(1,len(arr)):
  nodes.append(arr[i][0].lower())

for j in range(1,len(arr[0])):
  nodes.append(arr[0][j].lower())
print(nodes)
```

```python
heuristic = {}

for start in reversed(nodes):
  if start not in heuristic:
    heuristic[start] = {}
  dist = dijkstra(start,graph,nodes)
  for node in dist.keys():
    heuristic[start][node] = dist[node]-10
print(len(heuristic))
```
